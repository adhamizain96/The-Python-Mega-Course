```python
def sentence_maker(phrase):
    phrase_starters = ('how', 'who', 'what', 'where', 'when', 'why')
    phrase_capitalize = phrase.capitalize()
    if phrase.startswith(phrase_starters):
        return '{}?'.format(phrase_capitalize)
    else:
        return '{}.'.format(phrase_capitalize)
```

```python
sentence_maker('how are you doing today')
```

```python
sentence_maker('my name is Zainulabidin Adhami')
```

```python
results = []
while True:
    phrase = input('Say something: ')
    if phrase == '\end':
        break
    else:
        results.append(sentence_maker(phrase))
print(' '.join(results))
```
