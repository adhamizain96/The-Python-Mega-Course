```python
def calculate():
    math_fcn = input('Addition, Subtraction, Multiplication or Division: ')
    num_1 = float(input('Number 1: '))
    num_2 = float(input('Number 2: '))
    if math_fcn == 'Addition':
        print('Answer:',num_1 / num_2)
    elif math_fcn == 'Subtraction':   
        print('Answer:',num_1 - num_2)
    elif math_fcn == 'Multiplication':
        print('Answer:',num_1 * num_2)
    elif math_fcn == 'Division':
        print('Answer:',num_1 / num_2)
    else:
        print('Try Again!')
calculate()
```
